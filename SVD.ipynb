```python
import torch
import numpy as np
_ = torch.manual_seed(0)
```

```python
d, k = 10, 10

rank = 2

W = torch.randn(d, rank) @ torch.randn(rank, k)
print(W.shape)
print(W)
```

```python
rank = np.linalg.matrix_rank(W)
print(f'Rank of W: {rank}')
```

```python
U, S, V = torch.svd(W)

print(U.shape, S.shape, V.shape)
```
